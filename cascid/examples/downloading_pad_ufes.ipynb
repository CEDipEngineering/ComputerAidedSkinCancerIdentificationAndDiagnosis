# How to:

Download images from PAD-UFES.

### First Step:

Import cascid module's pad_ufes dataset.

```python
# Import isic downloading tools
from cascid.datasets.pad_ufes import database
```

### Second Step:

Use the 'get_df' function to download dataset if necessary. If the dataset is already downloaded, will read it immediately, if not, will attempt to download it automatically, and then load it.

```python
df = database.get_df()
df
```

And that's it! Now you have downloaded the PAD-UFES dataset.

### Preprocessing

Cascid has some preprocessing built-in. In order to use it, you may want to examine the specifics of each dataset, or simply use 'datasets.pipeline.preprocessing' to preprocess all your images.

```python
from cascid.datasets.pipeline.preprocessing import preprocess_dataset
preprocess_dataset('pad-ufes', 'all', image_shape=(512,512)) # Apply all available forms of preprocessing to pad-ufes, and save images in 512x512 RGB resolution.
```
