```python
import os
import cv2 as cv
import matplotlib.pyplot as plt
from cascid.configs.isic_cnf import IMAGES_DIR
from cascid.datasets.isic.database import get_df
```

```python
dataframe = get_df()
```

```python
dataframe.head()
```

```python
def get_importance(kp_element):
    return kp_element.response
```

```python
def apply_sift_surf(image_id):
    img = cv.imread(os.fspath(IMAGES_DIR) + "/" + image_id)
    im_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    resized = cv.resize(im_rgb, (256,256))
    surf = cv.SIFT_create()
    kp, des = surf.detectAndCompute(resized, None)
    list_importance = [point for point in kp]
    list_importance.sort(key=get_importance)
    list_rgb = list()
    for point in list_importance:
        point_coordinates = point.pt
        rgb = resized[int(point_coordinates[0])][int(point_coordinates[1])]
        list_rgb.append(rgb)
    print()
    return list_rgb
```

```python
dataframe["siftsurf"] = dataframe["img_id"].apply(lambda id: apply_sift_surf(id))
```
