```python
def verticalLine(plot, x, y, color, nome):
    plot.axvline(x, color=color, linestyle="dashed", linewidth=2)
    plot.text(1.01*x, y, nome, color=color)
```

### Tamanho

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
axes[0].hist(df.diameter_1, bins=30, color="orange", alpha=0.5, edgecolor="#1F8F50")
axes[0].hist(df.diameter_2, bins=30, color="turquoise", alpha=0.5, edgecolor="#1F8F50")
verticalLine(axes[0], df["diameter_1"].mean(), 300, "orange", f"MÃ©dia DiÃ¢metro 1: {df.diameter_1.mean():.2f}")
verticalLine(axes[0], df["diameter_2"].mean(), 200, "green", f"MÃ©dia DiÃ¢metro 2: {df.diameter_2.mean():.2f}")

axes[1].hist(df_melanoma.diameter_1, alpha=0.5, edgecolor="#1F8F50")
axes[1].hist(df_melanoma.diameter_2, color="indigo", alpha=0.5, edgecolor="#1F8F50")

verticalLine(axes[1], df_melanoma["diameter_1"].mean(), 10, "blue", f"MÃ©dia Melanoma DiÃ¢metro 1: { df_melanoma.diameter_1.mean():.2f}")
verticalLine(axes[1], df_melanoma["diameter_2"].mean(), 6, "purple", f"MÃ©dia Melanoma DiÃ¢metro 2: { df_melanoma.diameter_2.mean():.2f}")
```
